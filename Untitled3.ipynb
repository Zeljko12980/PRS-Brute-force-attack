```python
import numpy as np
from numba import cuda, uint32

@cuda.jit(device=True)
def left_rotate(n, b):
    return ((n << b) | (n >> (32 - b))) & 0xffffffff

@cuda.jit(device=True)
def sha1_transform(chunk, h):
    w = cuda.local.array(80, dtype=uint32)
    for i in range(16):
        w[i] = (
            (chunk[i * 4] << 24)
            | (chunk[i * 4 + 1] << 16)
            | (chunk[i * 4 + 2] << 8)
            | (chunk[i * 4 + 3])
        )
    for i in range(16, 80):
        w[i] = left_rotate(w[i-3] ^ w[i-8] ^ w[i-14] ^ w[i-16], 1)

    a, b, c, d, e = h[0], h[1], h[2], h[3], h[4]

    for i in range(80):
        if i < 20:
            f = (b & c) | ((~b) & d)
            k = 0x5A827999
        elif i < 40:
            f = b ^ c ^ d
            k = 0x6ED9EBA1
        elif i < 60:
            f = (b & c) | (b & d) | (c & d)
            k = 0x8F1BBCDC
        else:
            f = b ^ c ^ d
            k = 0xCA62C1D6

        temp = (left_rotate(a, 5) + f + e + k + w[i]) & 0xffffffff
        e = d
        d = c
        c = left_rotate(b, 30)
        b = a
        a = temp

    h[0] = (h[0] + a) & 0xffffffff
    h[1] = (h[1] + b) & 0xffffffff
    h[2] = (h[2] + c) & 0xffffffff
    h[3] = (h[3] + d) & 0xffffffff
    h[4] = (h[4] + e) & 0xffffffff

@cuda.jit
def sha1_kernel(messages, msg_lens, results):
    idx = cuda.grid(1)
    if idx >= messages.shape[0]:
        return

    # Initialize SHA-1 state
    h = cuda.local.array(5, dtype=uint32)
    h[0] = 0x67452301
    h[1] = 0xEFCDAB89
    h[2] = 0x98BADCFE
    h[3] = 0x10325476
    h[4] = 0xC3D2E1F0

    # Padding - we assume msg length < 56 for simplicity (one block)
    msg_len = msg_lens[idx]
    chunk = cuda.local.array(64, dtype=np.uint8)

    for i in range(64):
        if i < msg_len:
            chunk[i] = messages[idx, i]
        elif i == msg_len:
            chunk[i] = 0x80
        elif i >= 56:
            # Last 8 bytes for message length in bits
            chunk[i] = 0
        else:
            chunk[i] = 0

    # Insert message length in bits in last 8 bytes
    bit_len = msg_len * 8
    chunk[63] = bit_len & 0xFF
    chunk[62] = (bit_len >> 8) & 0xFF
    chunk[61] = (bit_len >> 16) & 0xFF
    chunk[60] = (bit_len >> 24) & 0xFF
    # chunk[59..56] = 0 for message length > 2^32 bits, which is fine for short msgs

    sha1_transform(chunk, h)

    # Write output
    for i in range(5):
        results[idx, i] = h[i]

# Test example
charset = "abc123"
password_len = 4
charset_bytes = np.array([ord(c) for c in charset], dtype=np.uint8)
charset_len = len(charset_bytes)

total_passwords = charset_len ** password_len

# Generate all possible passwords
results_pw = np.zeros((total_passwords, password_len), dtype=np.uint8)

threads_per_block = 256
blocks = (total_passwords + threads_per_block - 1) // threads_per_block

@cuda.jit
def generate_passwords(charset_bytes, charset_len, password_len, results):
    idx = cuda.grid(1)
    if idx >= results.shape[0]:
        return
    remainder = idx
    for i in range(password_len-1, -1, -1):
        results[idx, i] = charset_bytes[remainder % charset_len]
        remainder //= charset_len

generate_passwords[blocks, threads_per_block](charset_bytes, charset_len, password_len, results_pw)
cuda.synchronize()

# Prepare messages input for SHA1 kernel
# We pad the messages array to 64 bytes per message (max one block)
messages = np.zeros((total_passwords, 64), dtype=np.uint8)
msg_lens = np.full(total_passwords, password_len, dtype=np.int32)

for i in range(total_passwords):
    for j in range(password_len):
        messages[i, j] = results_pw[i, j]

# Allocate output for SHA1 digest (5 uint32 = 20 bytes)
results_sha1 = np.zeros((total_passwords, 5), dtype=np.uint32)

sha1_kernel[blocks, threads_per_block](messages, msg_lens, results_sha1)
cuda.synchronize()

# Convert first 10 SHA1 hashes to hex
for i in range(10):
    digest = results_sha1[i]
    hex_digest = ''.join(f'{(digest[j] >> (24 - 8*k)) & 0xFF:02x}' for j in range(5) for k in range(4))
    pwd = ''.join(chr(c) for c in results_pw[i])
    print(f"Password: {pwd}  SHA1: {hex_digest}")
```

```python
import hashlib
import hmac
import numpy as np
from numba import cuda

# Pretpostavljamo da je essid i ostalo poslano kao globalne varijable, ili preneseno na device
essid = b'YourSSIDHere'  # stavi pravi ESSID
mic = bytes.fromhex("3f0a57aa67231b5767f530c43db02948")  # primjer mic
mac_ap = bytes.fromhex("c4e984a611e8")
mac_cl = bytes.fromhex("0e098dc7ebfd")
nonce_ap = bytes.fromhex("4441524b53494445")
nonce_cl = bytes.fromhex("5d3e427fe1b61cbd0405a0e610936c355e99ed404")

charset = "abc123"
password_len = 8

# Prevedi charset u numpy array od ASCII kodova
charset_np = np.array([ord(c) for c in charset], dtype=np.uint8)

# Pomocna funkcija za PBKDF2 i HMAC nije trivialno prevesti u cuda,
# ovdje ćemo samo prikazati strukturu - kompleksne hash funkcije je najbolje raditi na CPU
# ili koristiti neku biblioteku za GPU hashing (OpenCL, CUDA libraries).

# Zato ćemo napraviti samo jednostavan primjer CUDA kernela koji prolazi kombinacije
# i šalje nazad "indeks" da bi ti mogao dalje provjeriti na CPU.

@cuda.jit
def generate_passwords(charset, length, results):
    idx = cuda.grid(1)
    total = 1
    for i in range(length):
        total *= len(charset)

    if idx >= total:
        return

    # dekodiranje indeksa u kombinaciju slova
    pw = cuda.local.array(8, dtype=np.uint8)  # maksimalna duzina 8
    temp = idx
    for i in range(length-1, -1, -1):
        pw[i] = charset[temp % len(charset)]
        temp //= len(charset)

    # Spremi generiranu lozinku u results (ovo je samo demonstracija)
    for i in range(length):
        results[idx, i] = pw[i]

# Podesi broj niti i blokova
threads_per_block = 256
num_blocks = (len(charset)**password_len + threads_per_block - 1) // threads_per_block

# Rezultati će biti matrica (broj_sifra, password_len)
results = np.zeros((len(charset)**password_len, password_len), dtype=np.uint8)

generate_passwords[num_blocks, threads_per_block](charset_np, password_len, results)

cuda.synchronize()

# Nakon ovoga možeš koristiti results na CPU za dalju proveru sa PBKDF2 i MIC

print("Generisane šifre (prvih 10):")
for i in range(min(10, len(results))):
    print("".join(chr(c) for c in results[i]))
```
