```python
import hashlib
import hmac
import string
import multiprocessing
from tqdm.notebook import tqdm
import torch
import ipywidgets as widgets
from IPython.display import display, clear_output

charset = string.ascii_letters + string.digits

def min_max(a, b):
    return (a, b) if a < b else (b, a)

def parse_hccapx(hashline):
    hl = hashline.strip().split("*")
    if len(hl) < 8:
        raise ValueError("Nevalidan format hc22000 linije")
    mic = bytes.fromhex(hl[2])
    mac_ap = bytes.fromhex(hl[3])
    mac_cl = bytes.fromhex(hl[4])
    essid = bytes.fromhex(hl[5])
    nonce_ap = bytes.fromhex(hl[6])
    nonce_cl = bytes.fromhex(hl[7][34:98])
    eapol_client = bytes.fromhex(hl[7])
    macs = min_max(mac_ap, mac_cl)
    nonces = min_max(nonce_ap, nonce_cl)
    ptk_inputs = b''.join([b'Pairwise key expansion\x00', macs[0], macs[1], nonces[0], nonces[1], b'\x00'])
    return mic, essid, ptk_inputs, eapol_client

def index_to_password(index, charset, length):
    base = len(charset)
    pwd = []
    for _ in range(length):
        pwd.append(charset[index % base])
        index //= base
    return ''.join(reversed(pwd))

def check_password_batch(args):
    # Ovo je worker funkcija za multiprocessing
    indices, mic, essid, ptk_inputs, eapol_client, password_len = args

    # Batch generisanje lozinki
    passwords = [index_to_password(idx, charset, password_len) for idx in indices]

    # PBKDF2 je na CPU (ovde koristi Python hashlib)
    for pwd in passwords:
        pmk = hashlib.pbkdf2_hmac('sha1', pwd.encode(), essid, 4096, 32)
        ptk = hmac.new(pmk, ptk_inputs, hashlib.sha1).digest()
        try_mic = hmac.new(ptk[:16], eapol_client, hashlib.sha1).digest()[:16]
        if try_mic == mic:
            return pwd
    return None

def brute_force_hybrid(hccapx_line, password_len, output_widget, batch_size=512):
    output_widget.clear_output()
    with output_widget:
        try:
            mic, essid, ptk_inputs, eapol_client = parse_hccapx(hccapx_line)
        except Exception as e:
            print(f"Gre≈°ka u unosu hc22000 linije: {e}")
            return

        total = len(charset) ** password_len
        cpu_count = multiprocessing.cpu_count()
        pool = multiprocessing.Pool(processes=cpu_count)

        print(f"Pokreƒáem brute-force sa du≈æinom lozinke = {password_len} i {cpu_count} CPU jezgara...")
        print(f"Ukupno poku≈°aja: {total}")

        # Pripremi indekse u batch-evima za multiprocessing
        batches = [range(i, min(i + batch_size, total)) for i in range(0, total, batch_size)]
        args_iter = ((batch, mic, essid, ptk_inputs, eapol_client, password_len) for batch in batches)

        pbar = tqdm(total=total)
        try:
            for result in pool.imap_unordered(check_password_batch, args_iter):
                pbar.update(batch_size)
                if result is not None:
                    print(f"\n‚úÖ Lozinka je pronaƒëena: {result}")
                    pool.terminate()
                    pool.join()
                    pbar.close()
                    return
            print("\n‚ùå Lozinka nije pronaƒëena.")
            pool.close()
            pool.join()
            pbar.close()
        except KeyboardInterrupt:
            print("\nPrekid od korisnika.")
            pool.terminate()
            pool.join()
            pbar.close()
        except Exception as e:
            print(f"\nDo≈°lo je do gre≈°ke: {e}")
            pool.terminate()
            pool.join()
            pbar.close()

# --- UI za Colab/Jupyter ---

hccapx_text = widgets.Textarea(
    value='',
    placeholder='Unesi hc22000 liniju ovde...',
    description='hc22000:',
    layout=widgets.Layout(width='90%', height='100px')
)

password_length_input = widgets.BoundedIntText(
    value=8,
    min=4,
    max=12,
    step=1,
    description='Du≈æina lozinke:',
)

button = widgets.Button(description="Pokreni brute-force")
output = widgets.Output()

def on_button_clicked(b):
    brute_force_hybrid(hccapx_text.value, password_length_input.value, output)

button.on_click(on_button_clicked)

display(hccapx_text, password_length_input, button, output)
```
